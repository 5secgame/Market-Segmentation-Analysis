```python
# Gender distribution visualization
labels = ['Female', 'Male']
sizes = df['Gender'].value_counts()
colors = ['red', 'yellow']
explode = [0, 0.1]
plt.figure(figsize=(5, 5))
plt.pie(sizes, colors=colors, explode=explode, labels=labels, shadow=True, autopct='%.2f%%')
plt.title('Gender Distribution', fontsize=20)
plt.axis('off')
plt.legend()
plt.show()
```

```python
# Customer segmentation based on psychographic segmentation
df['Like'] = df['Like'].replace({'I hate it!-5': '-5', 'I love it!+5': '+5'})
```

```python
# Visualizing Likelihood with respect to Age
sns.catplot(x="Like", y="Age", data=df, orient="v", height=6, aspect=2, palette="Set2", kind="swarm")
plt.title('Likelihood of McDonalds w.r.t Age')
plt.show()
```

```python
# Age distribution of customers
plt.figure(figsize=(25, 8))
f = sns.countplot(x=df['Age'], palette='hsv')
f.bar_label(f.containers[0])
plt.title('Age Distribution of Customers')
plt.show()
```

```python
# Encoding categorical variables
label_encoder = LabelEncoder()
for col in df.columns[:11]:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
# PCA
X = df.iloc[:, :11].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=11)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Explained variance
print("Explained variance by each principal component:")
print(pca.explained_variance_ratio_)
print("Cumulative explained variance:")
print(np.cumsum(pca.explained_variance_ratio_))
```

```python
# PCA loadings
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(11)], index=df.columns[:11])
plt.figure(figsize=(15, 10))
sns.heatmap(loadings, annot=True, cmap='Spectral')
plt.title('PCA Loadings')
plt.show()
```

```python
# Scree plot
plt.figure(figsize=(10, 5))
plt.plot(range(1, 12), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```
